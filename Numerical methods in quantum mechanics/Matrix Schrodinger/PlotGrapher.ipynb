```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy import interpolate as inter
import numpy as np
```

```python
h = 0.005
alpha = 1.0
a = 20.0
x = np.arange(0, a + h, h)
case = "alphaCase"
if case == "alphaCase":
    plt.figure(figsize=(8, 5))
    plt.title('График потенциала', fontsize = 20)
    plt.plot(x, alpha*x, color = 'red')
    plt.plot([0, 0], [0, 2 * a * alpha], color = 'red')
    plt.plot([a, a], [a * alpha, 2 * a * alpha], color = 'red')
    plt.show()
else:
    plt.figure(figsize=(8, 5))
    plt.title('График потенциала', fontsize = 20)
    plt.plot(x, alpha*x**2, color = 'red')
    plt.plot([0, 0], [0, 2 * a**2 * alpha], color = 'red')
    plt.plot([a, a], [a**2 * alpha, 2 * a**2 * alpha], color = 'red')
    plt.show()
plt.savefig('potentialGraph.jpg')
```

```python
x = np.array([], float)
psi = np.array([], float)

with open("plot.csv") as f:
    for line in f.readlines():
        x = np.append(x, float(line.split(' ')[0]))
        psi = np.append(psi, float(line.split(' ')[1]))
#plt.xlim(5, 10)
plt.ylim(-2, 2)
plt.plot(x*10, psi/100, linewidth = 2)
plt.savefig('Graph.jpg')
plt.show()
```
