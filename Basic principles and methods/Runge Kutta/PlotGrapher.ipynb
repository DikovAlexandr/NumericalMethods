```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy import interpolate as inter
import numpy as np
```

```python
fig, ax = plt.subplots()

fig.set_figheight(12)
fig.set_figwidth(20)

plt.rcParams['font.size'] = '20'

ax.set_title('Аналитического решения и решения методами Рунге-Кутты 2-го и 4-го и методом Адамса порядков',
             fontsize=20,
             loc='left',
             pad=10)
ax.set_ylabel('y',
              fontsize=20)
ax.set_xlabel('x',
              fontsize=20)
ax.grid(which='major',
        color='k')
ax.minorticks_on()
ax.grid(which='minor',
        color='gray',
        linestyle=':')

x1 = np.array([], float)
x2 = np.array([], float)
x3 = np.array([], float)
RK2y = np.array([], float)
RK4y = np.array([], float)
ADMy = np.array([], float)
y = np.array([], float)

with open("RK2.csv") as f:
    for line in f.readlines():
        x1 = np.append(x1, float(line.split(' ')[0]))
        RK2y = np.append(RK2y, float(line.split(' ')[1]))
        y = np.append(y, float(line.split(' ')[2]))

with open("RK4.csv") as f:
    for line in f.readlines():
        x2 = np.append(x2, float(line.split(' ')[0]))
        RK4y = np.append(RK4y, float(line.split(' ')[1]))

with open("Adams.csv") as f:
    for line in f.readlines():
        x3 = np.append(x3, float(line.split(' ')[0]))
        ADMy = np.append(ADMy, float(line.split(' ')[1]))

ax.plot(x1, y, linewidth=2, label='Аналитическое решение')
ax.plot(x1, RK2y, linewidth=2, label='Метод Рунге-Кутты 2-го порядка')
ax.plot(x2, RK4y, linewidth=2, label='Метод Рунге-Кутты 4-го порядка')
ax.plot(x3, ADMy, linewidth=2, label='Метод Адамса')

#plt.xlim(6, 8)

ax.scatter(x1, y, s=3)
ax.scatter(x1, RK2y, s=3)
ax.scatter(x2, RK4y, s=3)

# plot X - axis
ax.axhline(y=0, color='k')

# plot Y - axis
ax.axvline(x=0, color='k')
ax.legend()
plt.savefig('Graph.jpg')
plt.show()
```
