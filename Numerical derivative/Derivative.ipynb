```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy import interpolate as inter
import numpy as np
```

```python
fig, ax = plt.subplots()

fig.set_figheight(12)
fig.set_figwidth(20)

plt.rcParams['font.size'] = '20'

ax.set_title('Сравнение функции, её аналитической и численной производных',
          fontsize = 20,
          loc = 'left',
          pad = 10)
ax.set_ylabel('y',
           fontsize = 20)
ax.set_xlabel('x',
           fontsize = 20)
ax.grid(which='major',
        color = 'k')
ax.minorticks_on()
ax.grid(which='minor',
        color = 'gray',
        linestyle = ':')

x = np.array([], float)
f1 = np.array([], float)
f2 = np.array([], float)
f3 = np.array([], float)

with open("plotA.csv") as f:
  for line in f.readlines():
    x = np.append(x, float(line.split(' ')[0]))
    f1 = np.append(f1, float(line.split(' ')[1]))
    f2 = np.append(f2, float(line.split(' ')[2]))
    f3 = np.append(f3, float(line.split(' ')[2]))

ax.plot(x, f1, linewidth = 2, label='функция')
ax.plot(x, f2, linewidth = 2, label='аналитическая производная')
ax.plot(x, f3, linewidth = 2, label='численная производная')

ax.scatter(x, f1, s = 3)
ax.scatter(x, f2, s = 3)
ax.scatter(x, f3, s = 3)

# plot X - axis    
ax.axhline(y=0, color='k')

# plot Y - axis    
ax.axvline(x=0, color='k')

ax.legend()

plt.savefig('GraphA.jpg')

plt.show()
```
