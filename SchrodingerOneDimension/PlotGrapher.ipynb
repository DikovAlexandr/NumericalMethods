```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy import interpolate as inter
import numpy as np
```

```python
fig, ax = plt.subplots()

fig.set_figheight(12)
fig.set_figwidth(20)

plt.rcParams['font.size'] = '20'

ax.set_title('######################',
             fontsize=20,
             loc='left',
             pad=10)
ax.set_ylabel('y',
              fontsize=20)
ax.set_xlabel('x',
              fontsize=20)
ax.grid(which='major',
        color='k')
ax.minorticks_on()
ax.grid(which='minor',
        color='gray',
        linestyle=':')

x1 = np.array([], float)
x2 = np.array([], float)
y1 = np.array([], float)
y2 = np.array([], float)

with open("func11.txt") as f:
    for line in f.readlines():
        x1 = np.append(x1, float(line.split(' ')[0]))
        y1 = np.append(y1, float(line.split(' ')[1]))

with open("func12.txt") as f:
    for line in f.readlines():
        x2 = np.append(x2, float(line.split(' ')[0]))
        y2 = np.append(y2, float(line.split(' ')[1]))

ax.plot(x1, y1, linewidth=2, label='#')
ax.plot(x2, y2, linewidth=2, label='##')

#plt.xlim(6, 8)

ax.scatter(x1, y1, s=3)
ax.scatter(x2, y2, s=3)

# plot X - axis
ax.axhline(y=0, color='k')

# plot Y - axis
ax.axvline(x=0, color='k')
ax.legend()
#plt.savefig('Graph.jpg')
plt.show()
```
