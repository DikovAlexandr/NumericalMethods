```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy import interpolate as inter
import numpy as np
```

```python
fig, ax = plt.subplots()

fig.set_figheight(12)
fig.set_figwidth(20)

plt.rcParams['font.size'] = '20'

ax.set_title('Зависимость потенциала от модуля радиус вектора при разных значениях theta',
          fontsize = 20,
          loc = 'left',
          pad = 10)
ax.set_ylabel('phi',
           fontsize = 20)
ax.set_xlabel('r',
           fontsize = 20)
ax.grid(which='major',
        color = 'k')
ax.minorticks_on()
ax.grid(which='minor',
        color = 'gray',
        linestyle = ':')

x = np.array([], float)
f1 = np.array([], float)
f2 = np.array([], float)
f3 = np.array([], float)
f4 = np.array([], float)
f5 = np.array([], float)

with open("potential.csv") as f:
  for line in f.readlines():
    x = np.append(x, float(line.split(' ')[0]))
    f1 = np.append(f1, float(line.split(' ')[1]))
    f2 = np.append(f2, float(line.split(' ')[2]))
    f3 = np.append(f3, float(line.split(' ')[3]))
    f4 = np.append(f4, float(line.split(' ')[4]))
    f5 = np.append(f5, float(line.split(' ')[5]))

plt.xlim((None,20))
plt.ylim((0.05, 0.6))

ax.plot(x, f1, linewidth = 2, label='0')
ax.plot(x, f2, linewidth = 2, label='30')
ax.plot(x, f3, linewidth = 2, label='45')
ax.plot(x, f4, linewidth = 2, label='60')
ax.plot(x, f5, linewidth = 2, label='90')

ax.scatter(x, f1, s = 3)
ax.scatter(x, f2, s = 3)
ax.scatter(x, f3, s = 3)

# plot X - axis    
ax.axhline(y=0, color='k')

# plot Y - axis    
ax.axvline(x=0, color='k')

ax.legend()

plt.savefig('Graph.jpg')

plt.show()
```
