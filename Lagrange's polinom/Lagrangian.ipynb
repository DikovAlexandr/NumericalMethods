```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy import interpolate as inter
import numpy as np
```

```python
fig, ax = plt.subplots()

fig.set_figheight(12)
fig.set_figwidth(20)

ax.set_title('Сравнение функции и разложения в полином Лагранжа',
          fontsize = 14,
          loc = 'left',
          pad = 10)
ax.set_ylabel('y',
           fontsize = 14)
ax.set_xlabel('x',
           fontsize = 14)
ax.grid(which='major',
        color = 'k')
ax.minorticks_on()
ax.grid(which='minor',
        color = 'gray',
        linestyle = ':')

x = np.array([], float)
y = np.array([], float)
l = np.array([], float)

with open("plot.csv") as f:
  for line in f.readlines():
    x = np.append(x, float(line.split(' ')[0]))
    l = np.append(l, float(line.split(' ')[1]))
    y = np.append(y, float(line.split(' ')[2]))

ax.plot(x, y, linewidth = 3, label='функция')
ax.plot(x, l, linewidth = 3, label='лагранжиан')

ax.scatter(x, y, s = 10)
ax.scatter(x, l, s = 10)

# plot X - axis    
ax.axhline(y=0, color='k')

# plot Y - axis    
ax.axvline(x=0, color='k')

ax.legend()

plt.show()

plt.savefig('Graph.png')
```
